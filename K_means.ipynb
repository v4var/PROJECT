```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, Y = make_blobs(
    n_samples=1000,
    n_features=2,
    centers=5,
    cluster_std=1.0,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=20,
    return_centers=False,
)
```

```python
print(X.shape)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
```

```python
k = 5

color = ["red", "yellow", "blue", "green", "purple"]

clusters = {}

for i in range(k):
    center = np.random.randint(-12,12,2)
    points = []

    cluster = {
        'center' : center,
        'points' : points,
        'color' : color[i]
    }

    clusters[i] = cluster
```

```python
clusters
```

```python
def distance(k1, p):
    return np.sqrt(np.sum((k1-p)**2))
```

```python
def assignment_culsters(cluster):
    for i in range(X.shape[0]):
        dist = []

        for j in range(k):
            dist.append(distance(X[i], clusters[j]['center']))

        curr_cluster = np.argmin(dist)
        clusters[curr_cluster]['points'].append(X[i])
```

```python
def updation_clusters(clusters):
    for i in range(k):
        pts =  np.array(clusters[i]['points'])

        if pts.shape[0] > 0:
            u = np.mean(pts, axis = 0)
            clusters[i]['center'] = u
            clusters[i]['points'] = []
```

```python
def plot_clusters(clusters):
    for i in range(k):
        pts = np.array(clusters[i]['points'])

        try:
            plt.scatter(pts[:, 0], pts[:, 1], color = clusters[i]['color'])
        except:
            pass
        centroid = clusters[i]['center']
        plt.scatter(centroid[0], centroid[1], color = 'black', marker = '^')
```

```python
clusters
```

```python
assignment_culsters(clusters)
plot_clusters(clusters)
```

```python
assignment_culsters(clusters)
plot_clusters(clusters)
updation_clusters(clusters)
```
