```python
from urllib.request import urlopen
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area"
```

```python
data = urlopen(url)
```

```python
data_all = data.read()
```

```python
data_all
```

```python
from bs4 import BeautifulSoup
```

```python
data_all = BeautifulSoup(data_all, 'html')
```

```python
data_all.find_all('table', {})
```

```python
table = data_all.find('table', {'class' : 'wikitable sortable sticky-header col2left'})
```

```python
table
```

```python
table.find_all('th')[2].text
```

```python
heading = table.find_all('th')
```

```python
heading
```

```python
col_heading = []
for i in heading:
    col_heading.append(i.text[:-1])
```

```python
col_heading
```

```python
table_data = table.find_all('tr')
```

```python
table_data
```

```python
table_row = []

for i in table_data:
    row = i.find_all('td')
    curr_row = []
    for col_data in row:
        curr_row.append(col_data.text)
    table_row.append(curr_row)
print(table_row)
```

```python
table_row[9]
```

```python
csv_filename = 'scrapped_table.csv'
```

```python
a = ['a', 'b', 'c']
print(''.join(a))
print(','.join(a))
```

```python
import os
import pandas as pd
```

```python
with open(csv_filename, 'w', encoding = 'utf_8') as f:
    heading = ','.join(col_heading)
    heading += '\n'
    f.write(heading)

    for rows in table_row:
#        print(rows)
        temp_str = ','.join(rows)
        f.write(temp_str)
```

```python
csv_file_name = 'Scraped_Table.csv'

with open(csv_file_name, 'w', encoding = 'utf_8') as f:
    df = pd.DataFrame(table_row, columns = [col_heading])
    print(df)
    df.to_csv('pandas_scraping.csv')
```
